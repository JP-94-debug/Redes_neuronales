```python
# prompt: create a script to ask for an int number greater than 10, and then make the sum for the first n natural numbers, plot the answer

def sum_natural_numbers(n):
  """Calculates the sum of the first n natural numbers."""
  return (n * (n + 1)) // 2

while True:
  try:
    n = int(input("Enter an integer greater than 10: "))
    if n > 10:
      break
    else:
      print("Please enter a number greater than 10.")
  except ValueError:
    print("Invalid input. Please enter an integer.")

sum_of_numbers = sum_natural_numbers(n)
print(f"The sum of the first {n} natural numbers is: {sum_of_numbers}")

import matplotlib.pyplot as plt

# Create a simple plot
plt.plot([n], [sum_of_numbers], marker='o', linestyle='-', color='blue')
plt.xlabel("Number (n)")
plt.ylabel("Sum of Natural Numbers")
plt.title("Sum of First N Natural Numbers")
plt.grid(True)
plt.show()

```

```python
# prompt: create a program that recreates the famous problem of 3n+1

import matplotlib.pyplot as plt

def collatz_sequence(n):
  """Generates the Collatz sequence for a given number."""
  sequence = [n]
  while n != 1:
    if n % 2 == 0:
      n = n // 2
    else:
      n = 3 * n + 1
    sequence.append(n)
  return sequence

while True:
  try:
    n = int(input("Enter a positive integer: "))
    if n > 0:
      break
    else:
      print("Please enter a positive integer.")
  except ValueError:
    print("Invalid input. Please enter an integer.")

sequence = collatz_sequence(n)
print(f"The Collatz sequence for {n} is: {sequence}")

# Create a plot of the sequence
plt.plot(sequence, marker='o', linestyle='-', color='blue')
plt.xlabel("Step")
plt.ylabel("Value")
plt.title(f"Collatz Sequence for {n}")
plt.grid(True)
plt.show()
```
