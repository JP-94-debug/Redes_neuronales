```python
# prompt: create a script to ask for an int number greater than 10, and then make the sum for the first n natural numbers, plot the answer

def sum_natural_numbers(n):
  """Calculates the sum of the first n natural numbers."""
  return (n * (n + 1)) // 2

while True:
  try:
    n = int(input("Enter an integer greater than 10: "))
    if n > 10:
      break
    else:
      print("Please enter a number greater than 10.")
  except ValueError:
    print("Invalid input. Please enter an integer.")

sum_of_numbers = sum_natural_numbers(n)
print(f"The sum of the first {n} natural numbers is: {sum_of_numbers}")

import matplotlib.pyplot as plt

# Create a simple plot
plt.plot([n], [sum_of_numbers], marker='o', linestyle='-', color='blue')
plt.xlabel("Number (n)")
plt.ylabel("Sum of Natural Numbers")
plt.title("Sum of First N Natural Numbers")
plt.grid(True)
plt.show()

```
